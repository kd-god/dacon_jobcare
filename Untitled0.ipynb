```python
pivota_table=df.groupby(['SALEDATE','PUM_NM']).mean()
```

```python
pivota_table = pivota_table.reset_index()
```

```python
# pivota_table.tail(2)
```

```python
pivota_talbe = pivota_table.categorize(columns = 'PUM_NM')
```

```python
df_pivot = dd.pivot_table(pivota_talbe, index ='SALEDATE', columns ='PUM_NM', values='NEW_COST')
```
